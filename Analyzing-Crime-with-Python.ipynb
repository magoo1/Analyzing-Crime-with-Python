```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Crime_Data_from_2015_2.csv")
df.head()
```

```python
df['Date Occurred'] = pd.to_datetime(df['Date Occurred'],format= '%m/%d/%Y')
df.head()
```

```python
import calendar
```

```python
df['Month'] = (df['Date Occurred'].dt.month).apply(lambda x:calendar.month_abbr[x])
df.head()
```

```python
df['Month'] = pd.Categorical(df['Month'] , categories= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Nov', 'Dec'],ordered = True)
```

```python
df['Weekday'] =df['Date Occurred'].dt.weekday_name
df['Weekday'] = pd.Categorical(df['Weekday'], categories=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],ordered=True)
df.head()
```

```python
crimes = df
```

```python
crimes.groupby(['Month' , 'Weekday'])['DR Number'].count().unstack().plot(kind='bar')
plt.ylabel('No of crimes')
plt.show()
```

```python
crimes.groupby('Weekday')['DR Number'].count().plot(marker = 'o',label = 'crimes')
plt.ylabel('No of Crimes')
plt.show()
```

```python
sns.countplot(x = 'Crime Code Description', data= crimes, order = crimes['Crime Code Description'].value_counts().index)
plt.xticks(rotation = 'vertical')
plt.ylabel('No of Crimes')
plt.show()
```

```python
temp = crimes.groupby('Area Name')['Reporting District'].count().sort_values(ascending = False)
temp = temp[:10]
temp
```

```python
temp.plot(kind = 'bar',color = 'green')
plt.ylabel('No of Crimes')
plt.show()
```

```python
sns.countplot(x = 'Weapon Description', data=crimes)
plt.ylabel('No of crimes')
plt.show()
```
